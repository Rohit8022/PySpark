
- PySpark Dataframe
- Reading The Dataset
- Checking the Datatypes of the Column(Schema)
- Selecting Columns And Indexing
- Check Describe option similar to Pandas
- Adding Columns
- Dropping columns
- Renaming Columns

```python
#PySpark Dataframe
#1st way
df_spark=spark.read.option('header','true').csv('test1.csv',inferSchema=True)
```

```python
df_spark.show()
```

```python
### Check the schema
df_spark.printSchema()
```

```python
#2nd way
df_spark=spark.read.csv('test1.csv',header=True,inferSchema=True)
```

```python
df_spark
```

```python
# Check the schema
df_spark.printSchema()
```

```python
type(df_spark)
```

```python
df_spark.head(3)
```

```python
#Selecting Columns And Indexing
df_spark.select('name','age').show()
```

```python
df_spark.select(['name','Experience']).show()
```

```python
df_spark.dtypes
```

```python
df_spark.describe().show()
```

```python
# Adding Columns in data frame
df_spark=df_spark.withColumn('Salary after 2 years',df_spark['Experience']+2)
```

```python
## removing coloumn
df_spark=df_spark.drop('Salary after 2 years')
```

```python
df_spark.show()
```

```python
#Renaming Columns
df_spark=df_spark.withColumnRenamed('name','Names')
```

```python
df_spark.show()
```
