##Pyspark Handling Missing Values
- Dropping Columns
- Dropping Rows
- Various Parameter In Dropping functionalities
- Handling Missing values by Mean, MEdian And Mode

```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName('Practice').getOrCreate()
```

```python
df_spark=spark.read.csv('test2.csv',header=True,inferSchema=True)
```

```python
df_spark.show()
```

```python
#Dropping Columns
df_spark.drop('name').show()
```

```python
#Dropping row by null values
#how=any/All
df_spark.na.drop(how='all').show()# default value of how=any
```

```python
#thresh=
df_spark.na.drop(thresh=2).show()# should have atlest 2 non null values
```

```python
#subset=
df_spark.na.drop(subset=['age']).show()#will make col non-null by deleting all null value rows
```
