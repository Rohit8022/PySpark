```python
from pyspark.sql import SparkSession
```

```python
spark=SparkSession.builder.appName('Agg').getOrCreate()
df_spark=spark.read.csv('test3.csv',header=True,inferSchema=True)
```

```python
df_spark.show()
```

```python
# SUM
df_spark.groupBy('name').sum().show()
```

```python
df_spark.groupBy('departments','name').count().show()#can change name of agg function according to requirment
```

```python
# ANOTHER WAY OF WRITING THIS
df_spark.agg({'salary':'sum'}).show()
```
